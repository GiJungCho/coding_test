```python
def bottom_up(x):
    d = [0] * (x)

    for i in range(x):
    #     print(i)
        if i == 0 or i == 1:
            d[i] = 1

        else:
    #         print(i-2)
            d[i] = d[i-1] + d[i-2]

    print(d)
    return(d[x-1])

top_down(int(input()))
```

```python
# 탑다운
d = [0] * 100 
def pibo(x) :
    
    if x == 1 or x == 2 : 
        return 1 
    if d[x] != 0: 
        print( 'f('+ str(x) +')' , end= '' ) 
        return d[x] 
    d[x] = pibo(x - 1) + pibo(x - 2)
    return d[x]
```

```python
pibo(5)
```
